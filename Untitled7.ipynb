NOW TEXT PREPROCESSING

```python
# import the necessary libraries
import nltk
import string
import re
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
punc = '''!()0123456789-'[]{};:'"\,<>./?@#$%^&*_~'''
ps = PorterStemmer()
def preprocess(text):
  text=text.lower()
  word=nltk.word_tokenize(text)
  prep_data=[]
  for i in word:
    if i not in stop_words and i not in punc and i.isalpha():
      prep_data.append(ps.stem(i))
  return " ".join(prep_data) 
preprocess("I am gonna be home soon and i don't want 673 to talk about this stuff anymore tonight, k? I've cried enough today.")    

```

## here i have used port stemmer that will give root words make more easily the computaion and taking less space than lemmatization becuase in spam classifier stemming will work well .



```python
data['clear_msg'] = data['text'].apply(preprocess)
data.head()
```

Model Building

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
cv = CountVectorizer()
tfidf = TfidfVectorizer(max_features=3000)
```

```python
# x_train=tfidf.fit_transform(data['clear_msg']).toarray()
x_train=tfidf.fit_transform(data['clear_msg']).toarray()
y=data['label_num'].values
y
y.shape

```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test = train_test_split(x_train,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score
```

```python
gauss = GaussianNB()
multi = MultinomialNB()
bino = BernoulliNB()
gauss.fit(X_train,y_train)
y_pred1 = gauss.predict(x_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

# in spam classifier our objective should be decrease the false positive and increases the precision accuracy.

```python
multi.fit(X_train,y_train)
y_pred2 = multi.predict(x_test)
print(accuracy_score(y_test,y_pred2))
print(confusion_matrix(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
```

```python
bino.fit(X_train,y_train)
y_pred1 = bino.predict(x_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
gauss = GaussianNB()
multi = MultinomialNB()
bino = BernoulliNB()
svc = SVC(kernel='sigmoid', gamma=1.0)
knc = KNeighborsClassifier()
mnb = MultinomialNB()

clfs = {
    'SVC' : svc,
    'KNN' : knc, 
    'NB': mnb ,
    'gau':gauss
}

def best_classifier(model,X_train,y_train,X_test,y_test):
  model.fit(X_train,y_train)
  y_pred = model.predict(x_test)
  accuracy = accuracy_score(y_test,y_pred)
  precision = precision_score(y_test,y_pred)
  return accuracy,precision
```

```python
best_classifier(svc,X_train,y_train,x_test,y_test)
```

```python
ac_scores = []
precision_scores = []
for name,clf in clfs.items():
    current_accuracy,current_precision = best_classifier(clf, X_train,y_train,x_test,y_test)
    print("For ",name)
    print("Accuracy - ",current_accuracy)
    print("Precision - ",current_precision)
    ac_scores.append(current_accuracy)
    precision_scores.append(current_precision)
    
```
